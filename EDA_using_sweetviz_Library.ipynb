```python
import pandas as pd
df = pd.read_csv('/content/heart_attack_prediction_dataset.csv')
```

```python
df.head()
```

```python
import sweetviz as sv

# Analyzing the dataset
report = sv.analyze(df)

# Display the report
report.show_html("heart_attack_prediction_sweetviz_report.html")
```

```python
train = df.sample(frac=0.7, random_state=42)
test = df.drop(train.index)

report = sv.compare([train, "Training Data"], [test, "Test Data"])
report.show_html("train_vs_test_sweetviz_report.html")
```

```python
report = sv.analyze(df, target_feat="Heart Attack Risk")
report.show_html("target_analysis_sweetviz_report.html")
```

```python
import sweetviz as sv

# Specify feature configuration
feature_config = sv.FeatureConfig(skip="Patient ID", force_cat=["Continent"])

# Analyze the dataset with the feature configuration
report = sv.analyze(df, feat_cfg=feature_config)

# Display or save the report
report.show_html("feature_config_sweetviz_report.html")
```

```python
import sweetviz as sv

# Create subsets based on the "Sex" column
subset_male = df[df["Sex"] == "Male"]
subset_female = df[df["Sex"] == "Female"]

# Compare the two subsets
report = sv.compare([subset_male, "Male"], [subset_female, "Female"])

# Display or save the report
report.show_html("male_vs_female_sweetviz_report.html")
```
